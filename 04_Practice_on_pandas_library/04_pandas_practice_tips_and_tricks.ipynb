# 16- Splitting a string into multiple columns

```python
# import libraries
import pandas as pd


df = pd.DataFrame({'name':['Ahmad Bhutta', 'Ali Bhutta', 'Sara Khan'],
                'location':['Lahore Pakistan', 'Karachi Pakistan', 'Islamabad Pakistan'],})
df
```

```python
df.name.str.split(' ', expand=True)
```

```python
# adding those splits into new columns
df[["first_name", "last_name"]] = df.name.str.split(' ', expand=True)
df
```

```python
df[["city", "country"]] = df.location.str.split(' ', n=1, expand=True)
df
```

```python
# Refine data manipulation
df = df[['first_name', 'last_name', 'city', 'country']] 
df
```
